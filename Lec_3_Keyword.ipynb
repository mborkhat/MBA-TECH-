# What is encryption?

Encryption is the method by which information is converted into secret code that hides the information's true meaning. The science of encrypting and decrypting information is called cryptography.

In computing, unencrypted data is also known as plaintext, and encrypted data is called ciphertext. 

encrypt.png


*   Cipher :  The formulas used to encode and decode messages are called encryption algorithms
*   Key : The variable, which is called a key, is what makes a cipher's output unique

Historically, it was used by militaries and governments. 

In modern times, encryption is used to protect data stored on computers and storage devices, as well as data in transit over networks.

## List comprehension
List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Example:

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.

Without list comprehension you will have to write a for statement with a conditional test inside:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango","apple"]
newlist = []

for x in fruits:
  if "apple" in x:
    newlist.append(x)

print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango","apple"]
newlist = []

for x in fruits:
  if "apple" not in x:
    newlist.append(x)

print(newlist)
```

```python
newlist = [x for x in fruits if x in "apple"]
print(newlist)
```

Repeats of letters in the word are removed, then the cipher alphabet is generated with the keyword matching to A, B, C, etc. until the keyword is used up, whereupon the rest of the ciphertext letters are used in alphabetical order, excluding those already used in the key.

### **FINAL ENCRYPTION FUNCTION**

```python
# Python program for encoding the string
message = "Attack at dawn"
keyword = "Star War"
ciphertext = ""
```

```python
# removes default elements
# remove duplicates
def duplicates(list):
  key = []
  for i in list:
    if i not in key:
      key.append(i)
        
  return key
```

```python
def keyword_encrypt(keyword,message):
  # Python program for encoding the string
  ciphertext = ""

  # stores all upper case alphabets
  all_alphabets = []
  for i in range(97,123):
    all_alphabets.append(chr(i).upper())

  # converts message to list
  msg = []
  for i in message:
      msg.append(i.upper())
      
  #msg = [x.upper() for x in message]

  keyword1 = keyword.upper()

  keyword1 = duplicates(keyword1)
  #print("keyword after removing duplicate",keyword1)

  # Stores the encryption list
  encrypting = duplicates(keyword1+all_alphabets)
  #print("Enctyptig text is",encrypting)

  #removes spaces from the encryption list
  for i in encrypting:
    if(i == ' '):
      encrypting.remove(' ')

  # maps each element of the message to the encryption list and stores it in ciphertext
  for i in range(len(msg)):
    if(msg[i] != ' '):
      ciphertext = ciphertext+encrypting[all_alphabets.index(msg[i])]
    else:
      ciphertext = ciphertext+' '

  return ciphertext
```

```python
print("Keyword : ", keyword)
print("Message before Ciphering : ", message)
print("Ciphered Text : ", keyword_encrypt(keyword,message))
```

# Final Code Decryption

```python
# Python program for encoding the string
message = ""
keyword = "Star War"
ciphertext = "SPPSAG SP RSVJ"
```

```python
def duplicates(list):
	key = []
	for i in list:
		if i not in key:
			key.append(i)

	return key
```

```python
def keyword_decrypt(keyword,message):
  # Python program for encoding the string
  message = ""

  # stores all upper case alphabets
  all_alphabets = []
  for i in range(65,91):
    all_alphabets.append(chr(i))
 

  # converts message to list
  ct = []
  for i in ciphertext:
      ct.append(i.upper())
      
  #msg = [x.upper() for x in message]

  keyword1 = keyword.upper()

  keyword1 = duplicates(keyword1)
 # print("keyword after removing duplicate",keyword1)

  # Stores the encryption list
  encrypting = duplicates(keyword1+all_alphabets)
  #print("Enctyptig text is",encrypting)

  #removes spaces from the encryption list
  for i in encrypting:
    if(i == ' '):
      encrypting.remove(' ')

  # maps each element of the message to the encryption list and stores it in ciphertext
  for i in range(len(ct)):
    if(ct[i] != ' '):
      message = message+all_alphabets[encrypting.index(ct[i])]
    else:
      message = message+' '

  return message
```
